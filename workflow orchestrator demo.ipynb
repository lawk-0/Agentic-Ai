```python
# WAYWARD WORLD – Workflow vs Orchestrator Demonstration
# Single-cell simulation with animation, monitoring, adaptation, and replanning

import random
import time
from IPython.display import clear_output
from collections import deque

GRID = 6
start = (0,0)
goal = (5,5)

# initial obstacles
obstacles = {(2,2),(3,3)}

moves = [(1,0),(-1,0),(0,1),(0,-1)]

# -----------------------------
# GRID VISUALIZATION
# -----------------------------

def draw_grid(agent, obstacles):

    grid = []

    for i in range(GRID):
        row = []
        for j in range(GRID):

            if (i,j) == agent:
                row.append("A")

            elif (i,j) == start:
                row.append("S")

            elif (i,j) == goal:
                row.append("G")

            elif (i,j) in obstacles:
                row.append("X")

            else:
                row.append(".")

            row

        grid.append(row)

    for r in grid:
        print(" ".join(r))


# -----------------------------
# PLANNER
# -----------------------------

def planner(start, goal, obstacles):

    queue = deque([(start, [])])
    visited = set()

    while queue:

        pos, path = queue.popleft()

        if pos == goal:
            return path

        if pos in visited:
            continue

        visited.add(pos)

        for dx,dy in moves:

            nx,ny = pos[0]+dx, pos[1]+dy

            if 0 <= nx < GRID and 0 <= ny < GRID:

                if (nx,ny) not in obstacles:

                    queue.append(((nx,ny), path+[(nx,ny)]))

    return None


# -----------------------------
# ENVIRONMENT MONITOR
# -----------------------------

def monitor_environment(obstacles):

    if random.random() < 0.30:

        new_obstacle = (
            random.randint(0,GRID-1),
            random.randint(0,GRID-1)
        )

        if new_obstacle != start and new_obstacle != goal:

            obstacles.add(new_obstacle)

            print("\n[ENVIRONMENT CHANGE]")
            print("New obstacle appeared at:", new_obstacle)

    return obstacles


# -----------------------------
# ADAPTATION CHECK
# -----------------------------

def adaptation_possible(agent, obstacles):

    for dx,dy in moves:

        nx,ny = agent[0]+dx, agent[1]+dy

        if 0 <= nx < GRID and 0 <= ny < GRID:

            if (nx,ny) not in obstacles:

                return (nx,ny)

    return None


# -----------------------------
# ORCHESTRATOR
# -----------------------------

def orchestrator(start, goal, obstacles):

    agent = start

    print("\n===== WAYWARD WORLD SIMULATION =====\n")

    while agent != goal:

        print("\n[ORCHESTRATOR] Planning workflow from", agent)

        path = planner(agent, goal, obstacles)

        if not path:

            print("[ORCHESTRATOR] No valid path exists")
            return

        print("\n[WORKFLOW GENERATED]")
        print("Planned path:", path)

        for step in path:

            clear_output(wait=True)

            print("\n--- ENVIRONMENT MONITORING ---")

            obstacles = monitor_environment(obstacles)

            if step in obstacles:

                print("\n[ORCHESTRATOR ALERT]")
                print("Obstacle detected at planned step:", step)

                alt = adaptation_possible(agent, obstacles)

                if alt:

                    print("\n[ORCHESTRATOR DECISION]")
                    print("Choosing ADAPTATION (local adjustment)")

                    agent = alt

                    print("Agent adapted to:", agent)

                else:

                    print("\n[ORCHESTRATOR DECISION]")
                    print("REPLANNING required")

                break

            agent = step

            print("\nAgent moved to:", agent)

            draw_grid(agent, obstacles)

            time.sleep(0.8)

            if agent == goal:

                print("\n===== GOAL REACHED =====")

                return


# -----------------------------
# RUN DEMO
# -----------------------------

orchestrator(start, goal, obstacles)
```
