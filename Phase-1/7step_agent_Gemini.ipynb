```python
# ==========================================
# Gemini Agent Architecture Generator
# Produces 7-step architecture in bullet points
# ==========================================

import google.generativeai as genai

# ------------------------------------------
# Configure API key
# ------------------------------------------

genai.configure(api_key="YOUR_GEMINI_API_KEY")

model = genai.GenerativeModel("gemini-2.5-flash")

# ------------------------------------------
# User input
# ------------------------------------------

system_name = input("Enter a system or activity: ")

# ------------------------------------------
# Prompt
# ------------------------------------------

prompt = f"""
You are an AI systems architect.

For the system: {system_name}

Generate the 7-step autonomous agent architecture using bullet points only.

Steps:

1. Goal Interpretation
2. Task Decomposition
3. Task Sequencing / Planning
4. Agent Architecture
5. Execution
6. Execution Monitoring
7. Replanning / Adaptation

Rules:
- Each step must contain 2-4 bullet points
- Each bullet point must be short (max 10 words)
- No paragraphs
"""

# ------------------------------------------
# Generate response
# ------------------------------------------

response = model.generate_content(prompt)

print("\nAgent Planning Architecture:\n")
print(response.text)
```
