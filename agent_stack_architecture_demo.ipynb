# Agent Stack Architecture Demo
This notebook demonstrates a simplified **Agent Stack Architecture**:

LLM → Planner → Orchestrator → Tools → Environment

Task Example: Debugging and improving a Python program.

## 1. Install and Import Libraries

```python
import os
from openai import OpenAI

# Set your OpenAI API key
os.environ['OPENAI_API_KEY'] ='KEY I YOUR API'

client = OpenAI()
```

## 2. Environment
The environment contains the program that needs debugging.

```python
buggy_program = '''
def average(numbers):
    total = 0
    for i in range(len(numbers)):
        total += i
    return total / len(numbers)

print(average([10,20,30]))
'''

print(buggy_program)
```

## 3. LLM Reasoning Layer
The LLM analyzes the problem and suggests fixes.

```python
def call_llm(prompt):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )
    return response.choices[0].message.content
```

## 4. Planner Layer
The planner decomposes the debugging goal into tasks.

```python
def planner(code):
    prompt = f"""
You are a planning agent.
Break the debugging process into steps.

Code:
{code}
"""
    return call_llm(prompt)
```

## 5. Tools Layer
Tools perform actions like executing code or modifying code.

```python
def run_code(code):
    try:
        exec(code)
        return "Code executed successfully"
    except Exception as e:
        return str(e)
```

## 6. Debug Tool Using LLM

```python
def debug_code(code):
    prompt = f"""
You are a software debugging assistant.
Identify the bug and provide corrected code.

Code:
{code}
"""
    return call_llm(prompt)
```

## 7. Orchestrator Layer
Coordinates planner, tools, and LLM.

```python
def orchestrator(code):
    print("--- Planning Steps ---")
    plan = planner(code)
    print(plan)

    print("\n--- Running Code ---")
    result = run_code(code)
    print(result)

    if "error" in result.lower() or "exception" in result.lower():
        print("\n--- Debugging ---")
        fix = debug_code(code)
        print(fix)
        return fix

    return "No errors detected"
```

## 8. Run the Agent System

```python
orchestrator(buggy_program)
```
