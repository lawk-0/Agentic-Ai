Rational constraint becomes:
If temperature = mild → pressure = low
Not off — but low
Rain/Snow still → pressure = off

Rational Environmental Model
Pressure = f ( Weather, Temperature)
Pressure = Off if Weather is snow or Rain
Pressure  = High if temperature is warm or hot
Pressure  = low if temperature is mild

We will have max 2 times per day to keep sprinkler on
If temperature is Hot or warm then sprinker will be switched on twice a day
If temperature is mild and the sprinkler was on for 20 minutes then we can not open the sprinkler second time


This agent will randomly select temperature while user provides input for weather for the day
This following code is for reflex based rational agent

```python
import random

weather_states = ["rain", "snow", "clear", "humid"]

def select_temperature(weather):
    if weather == "rain":
        return random.choice(["cool", "mild"])
    elif weather == "snow":
        return "cold"
    elif weather == "clear":
        return random.choice(["mild", "warm", "hot"])
    elif weather == "humid":
        return random.choice(["humid", "warm"])

def run_simulation():

    print("\nEnter weather for 5 days (rain/snow/clear/humid)\n")

    daily_inputs = []
    for i in range(5):
        w = input(f"Day {i+1} Weather: ").lower()
        while w not in weather_states:
            print("Invalid input.")
            w = input(f"Day {i+1} Weather: ").lower()
        daily_inputs.append(w)

    print("\n--- Simulation Output ---\n")

    for day in range(5):

        print(f"=========== DAY {day+1} ===========")

        weather = daily_inputs[day]
        temperature = select_temperature(weather)

        print(f"Weather      : {weather}")
        print(f"Temperature  : {temperature}\n")

        # -----------------------------
        # RAIN / SNOW
        # -----------------------------
        if weather in ["rain", "snow"]:
            print("Cycle 1:")
            print("Pressure       : off")
            print("Soil Moisture  : wet")
            print("Sprinkler      : off")
            print("Irrigation Time: 0 minutes")
            print("-----------------------------------\n")
            continue

        soil_moisture = "dry"

        # -----------------------------
        # HOT DAY
        # -----------------------------
        if temperature == "hot":

            for cycle in range(2):
                print(f"Cycle {cycle+1}:")
                print("Pressure       : high")
                print("Soil Moisture  : wet")
                print("Sprinkler      : on")
                print("Irrigation Time: 20 minutes")
                print("-----------------------------------")
            print()
            continue

        # -----------------------------
        # HUMID DAY
        # -----------------------------
        if weather == "humid":

            # Cycle 1
            print("Cycle 1:")
            print("Pressure       : high")
            print("Soil Moisture  : wet")
            print("Sprinkler      : on")
            print("Irrigation Time: 20 minutes")
            print("-----------------------------------")

            # Cycle 2
            print("Cycle 2:")
            print("Pressure       : low")
            print("Soil Moisture  : wet")
            print("Sprinkler      : on")
            print("Irrigation Time: 10 minutes")
            print("-----------------------------------\n")
            continue

        # -----------------------------
        # MILD DAY
        # -----------------------------
        if temperature == "mild":

            print("Cycle 1:")
            print("Pressure       : low")
            print("Soil Moisture  : wet")
            print("Sprinkler      : on")
            print("Irrigation Time: 20 minutes")
            print("-----------------------------------")
            print("Cycle 2 skipped (mild day sufficient irrigation)\n")
            continue

        # -----------------------------
        # WARM DAY
        # -----------------------------
        if temperature == "warm":

            # Cycle 1
            print("Cycle 1:")
            print("Pressure       : high")
            print("Soil Moisture  : wet")
            print("Sprinkler      : on")
            print("Irrigation Time: 20 minutes")
            print("-----------------------------------")

            # Cycle 2 (maintenance)
            print("Cycle 2:")
            print("Pressure       : high")
            print("Soil Moisture  : wet")
            print("Sprinkler      : on")
            print("Irrigation Time: 10 minutes")
            print("-----------------------------------\n")
            continue

# Run simulation
run_simulation()
```
