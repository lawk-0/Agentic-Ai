```python
from smolagents import CodeAgent, DuckDuckGoSearchTool,LiteLLMModel, tool
import os
import litellm
litellm.suppress_debug_info = True

@tool
def celToFarenAndViceVersa(temp: float, unit: str) -> str:
    """
    Convert temperatures between Celsius and Fahrenheit.

    Args:
        temp (float): The temperature value to convert.
        unit (str): The unit of the given temperature. Use 'C' for Celsius or 'F' for Fahrenheit.

    Returns:
        str: Converted temperature string.

    Examples:
        Convert 25°C to Fahrenheit -> "25°C is equal to 77°F"
        Convert 77°F to Celsius -> "77°F is equal to 25°C"
    """
    if(unit.lower() == "c"):
        return f"{temp}°C is equal to {(temp * 9/5) + 32}°F"
    elif(unit.lower() == "f"):
        return f"{temp}°F is equal to {(temp - 32) * 5/9}°C"
    else:
        return "Invalid unit. Please use 'C' for Celsius or 'F' for Fahrenheit."
    
@tool
def randomMotivationQuotes() -> list:
    """
        Get random motivational quotes.

        Returns:
        list: A Random motivational quote from this list.
    """
    quotes = [
        "Believe you can and you're halfway there. - Theodore Roosevelt",
        "The only way to do great work is to love what you do. - Steve Jobs",
        "Don't watch the clock; do what it does. Keep going. - Sam Levenson",
        "The future belongs to those who believe in the beauty of their dreams. - Eleanor Roosevelt",
        "It does not matter how slowly you go as long as you do not stop. - Confucius"
    ]
    return quotes

@tool
def checkPrimeNumber(num: int) -> str:
    """
    Check if a number is prime.

    Args:
        num (int): The number to check.

    Returns:
        str: Whether the number is prime or not.
    """
    if num <= 1:
        return f"{num} is not a prime number."
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return f"{num} is not a prime number."
    return f"{num} is a prime number."

@tool
def reverseString(s: str) -> str:
    """
    Reverse a string.

    Args:
        s (str): The string to reverse.

    Returns:
        str: The reversed string.
    """
    return s[::-1]

groq_model = LiteLLMModel(
    model_id="groq/llama-3.1-8b-instant",
    api_key=os.getenv("GROQ_API_KEY")
)

agent = CodeAgent(
    tools=[DuckDuckGoSearchTool(), celToFarenAndViceVersa, randomMotivationQuotes, checkPrimeNumber, reverseString],
    model=groq_model
)

```

```python

print("\n--- Query 1: Celsius to Fahrenheit ---")
result1 = agent.run("Convert 25°C to Fahrenheit.")
print(result1)
```

```python
print("\n--- Query 2: Fahrenheit to Celsius ---")
result2 = agent.run("Convert 77°F to Celsius.")
print(result2)
```

```python
print("\n--- Query 3: Random Motivation Quotes ---")
result3 = agent.run("Give me a random motivation quote.")
print(result3)
```

```python
print("\n--- Query 4: Reverse a String ---")
result4 = agent.run("Reverse the string 'Hello, World!'")
print(result4)
```

```python
print("\n--- Query 5: Check if a Number is Prime ---")
result5 = agent.run("Is 17 a prime number?")
print(result5)
```
