```python
import random
import copy

GRID = 6

start = (0,0)
goal = (5,5)

# ----------------------------------
# Generate plan
# ----------------------------------

def generate_plan():

    path=[start]
    r,c=start

    while r<goal[0] or c<goal[1]:

        if c<goal[1]:
            c+=1
        elif r<goal[0]:
            r+=1

        path.append((r,c))

    return path

plan = generate_plan()

# ----------------------------------
# Grid display
# ----------------------------------

def show_grid(obstacles):

    grid=[["." for _ in range(GRID)] for _ in range(GRID)]

    grid[start[0]][start[1]]="S"
    grid[goal[0]][goal[1]]="G"

    for o in obstacles:
        grid[o[0]][o[1]]="X"

    for row in grid:
        print(row)

# ----------------------------------
# Generate obstacles
# ----------------------------------

def generate_obstacles(level):

    obstacles=[]

    if level=="free":
        return obstacles

    if level=="few":
        count=random.randint(1,2)

    if level=="many":
        count=random.randint(4,5)

    while len(obstacles)<count:

        r=random.randint(0,GRID-1)
        c=random.randint(0,GRID-1)

        if (r,c)!=start and (r,c)!=goal:
            obstacles.append((r,c))

    return obstacles

# ----------------------------------
# Monitoring decides execution strategy
# ----------------------------------

def choose_strategy(obstacles):

    if len(obstacles)==0:
        return "Sequential Execution"

    if len(obstacles)<=2:
        return "Reactive Execution"

    return "Adaptive Execution"

# ----------------------------------
# Execute plan
# ----------------------------------

def execute(plan,obstacles):

    strategy=choose_strategy(obstacles)

    print("\nEnvironment Grid:\n")
    show_grid(obstacles)

    print("\nExecution Strategy:",strategy,"\n")

    position=start

    for step in plan[1:]:

        print("Drone attempting:",step)

        # monitoring check
        if step in obstacles:

            if strategy=="Sequential Execution":
                print("Collision detected → sequential plan fails")
                break

            if strategy=="Reactive Execution":

                print("Obstacle detected → reactive adjustment")

                alt=(step[0]+1,step[1])

                if alt[0]<GRID:
                    step=alt

                print("Drone rerouted to:",step)

            if strategy=="Adaptive Execution":

                print("Adaptive system replanning")

                alt=(position[0]+1,position[1])

                if alt[0]>=GRID:
                    alt=(position[0],position[1]+1)

                step=alt

                print("New replanned step:",step)

        position=step

        print("Drone moved to:",position)

# ----------------------------------
# Run scenarios
# ----------------------------------

print("===== Scenario 1: Free Environment =====")
execute(plan,generate_obstacles("free"))

print("\n\n===== Scenario 2: Few Obstacles =====")
execute(plan,generate_obstacles("few"))

print("\n\n===== Scenario 3: Dense Environment =====")
execute(plan,generate_obstacles("many"))
```
